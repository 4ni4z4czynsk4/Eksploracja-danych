# Eksploracyjna Analiza Danych (EDA) - Zbiór Danych Sprzedaży Domów

## Krok 1: Ładowanie i przegląd danych

### Cel:
Załadowanie danych treningowych i testowych oraz przeprowadzenie wstępnego przeglądu struktury, typów danych i statystyk opisowych.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Konfiguracja stylu wykresów
sns.set_style("whitegrid")
plt.style.use('seaborn-v0_8-pastel') # Zmiana palety kolorów
plt.rcParams['figure.figsize'] = (12, 7) # Nieco większy rozmiar wykresów

# Wczytanie zbiorów danych
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')

print("### Informacje o zbiorze treningowym:")
print(f"Liczba wierszy i kolumn: {df_train.shape}")
print("\nPierwsze wiersze:")
display(df_train.head())

print("\nTypy danych kolumn:")
print(df_train.dtypes.value_counts())

print("\nPodstawowe statystyki opisowe (zmienne numeryczne):")
display(df_train.describe().T) # Transpozycja dla lepszej czytelności
```

## Krok 2: Analiza Zmiennej Docelowej - `SalePrice`

### Cel:
Szczegółowa analiza rozkładu zmiennej `SalePrice` (ceny sprzedaży), w tym statystyki, wizualizacja rozkładu i identyfikacja wartości odstających.

```python
target_variable = 'SalePrice'

print(f"### Analiza zmiennej docelowej: {target_variable}")
print("\nStatystyki opisowe:")
print(df_train[target_variable].describe())

print(f"\nMediana {target_variable}: {df_train[target_variable].median():,.2f}")
print(f"Średnia {target_variable}: {df_train[target_variable].mean():,.2f}")


# Wizualizacja rozkładu za pomocą histogramu i wykresu gęstości (KDE)
plt.figure(figsize=(12, 6))
sns.histplot(df_train[target_variable], bins=40, kde=True, color='skyblue') # Zmiana liczby bins i koloru
plt.title(f"Rozkład zmiennej {target_variable}", fontsize=16)
plt.xlabel("Cena Sprzedaży", fontsize=12)
plt.ylabel("Częstość", fontsize=12)
plt.show()

# Boxplot do identyfikacji outlierów
plt.figure(figsize=(12, 4))
sns.boxplot(x=df_train[target_variable], color='salmon') # Zmiana koloru
plt.title(f"Boxplot zmiennej {target_variable}", fontsize=16)
plt.xlabel("Cena Sprzedaży", fontsize=12)
plt.show()

# Obliczenie i interpretacja skośności
skewness = df_train[target_variable].skew()
print(f"\nSkośność rozkładu {target_variable}: {skewness:.3f}") # Większa precyzja
if abs(skewness) > 1:
    print("Wnioski: Rozkład jest silnie skośny.")
elif abs(skewness) > 0.5:
    print("Wnioski: Rozkład jest umiarkowanie skośny.")
else:
    print("Wnioski: Rozkład jest w przybliżeniu symetryczny.")

print("\nInterpretacja: Silna prawoskośność sugeruje, że większość domów ma niższe ceny, a mniejsza liczba domów osiąga znacznie wyższe ceny. Obecność outlierów na boxplocie potwierdza występowanie bardzo drogich nieruchomości.")
```

## Krok 3: Zależność między Powierzchnią a Ceną

### Cel:
Zbadanie, w jaki sposób różne zmienne związane z powierzchnią wpływają na cenę sprzedaży. Analiza korelacji i wizualizacja zależności za pomocą wykresów punktowych.

```python
import re

# Wyszukanie kolumn związanych z powierzchnią (ignorując wielkość liter)
area_features = [col for col in df_train.columns if re.search(r'Area|SF', col, re.IGNORECASE)]
print("### Zmienne związane z powierzchnią:")
print(area_features)

# Obliczenie korelacji z ceną sprzedaży i sortowanie
correlations_with_price = df_train[area_features + [target_variable]].corr()[target_variable].sort_values(ascending=False)
print(f"\n### Korelacja zmiennych powierzchniowych z {target_variable}:")
display(correlations_with_price)

# Wybór kilku kluczowych zmiennych powierzchni do wizualizacji
key_area_features = ['GrLivArea', 'TotalBsmtSF', '1stFlrSF', 'GarageArea', 'LotArea']

# Tworzenie wykresów punktowych (scatter plots)
num_plots = len(key_area_features)
num_cols = 2
num_rows = (num_plots + num_cols - 1) // num_cols

plt.figure(figsize=(num_cols * 6, num_rows * 5)) # Mniejszy rozmiar sub-wykresów

for i, col in enumerate(key_area_features):
    plt.subplot(num_rows, num_cols, i + 1)
    sns.scatterplot(data=df_train, x=col, y=target_variable, alpha=0.6, color=sns.color_palette("viridis")[i % len(sns.color_palette("viridis"))]) # Zmiana palety kolorów
    plt.title(f'{col} vs {target_variable}')
    plt.xlabel(col)
    plt.ylabel(target_variable)

plt.tight_layout()
plt.show()

print("\n### Obserwacje z wykresów zależności powierzchni od ceny:")
print("- Zmienne takie jak 'GrLivArea', 'TotalBsmtSF', '1stFlrSF' i 'GarageArea' wykazują wyraźną, pozytywną zależność liniową z ceną.")
print("- 'LotArea' ma słabszą i bardziej rozproszoną zależność.")
print("- Na wykresach dla powierzchni mieszkalnych i garażu widoczne są punkty odstające, które mogą wymagać dalszej analizy.")
```
